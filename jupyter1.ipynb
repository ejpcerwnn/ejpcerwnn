<span style="color:blue"> AFFICHER un DATA FRAME </span>

```python
import pandas as pd

d = {'col1': [1, 2, 3, 4, 7], 'col2': [4, 5, 6, 9, 5], 'col3': [7, 8, 12, 1, 11]}

df = pd.DataFrame(data=d)

print(df)
```

<span style="color:blue"> AFFICHAGE du nombre des COLONNES de DATA FRAME </span>

```python
count_column = df.shape[1]
print(count_column)
```

<span style="color:blue"> AFFICHAGE NOMBRE DE LIGNES de DATA FRAME </span>

```python
count_row = df.shape[0]
print(count_row)
```

<span style="color:blue"> MOYENNE d'un tableau </span>

```python
import numpy as np

Calorie_burnage = [240, 250, 260, 270, 280, 290, 300, 310, 320, 330]

Average_calorie_burnage = np.mean(Calorie_burnage)

print(Average_calorie_burnage)
```
